```python

# 3. Data inspection
print("\nProducts info:")
print(products.info())
print("\nOrders info:")
print(orders.info())
```

```python

# 4. Merge Orders and Products
merged = orders.merge(products, on='product_id')
print("\nMerged dataset:")
print(merged.head())

```

```python
# 5. Sales by category
sales_by_category = merged.groupby('product_category')['price'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x=sales_by_category.index, y=sales_by_category.values, color="steelblue")

plt.title('Omsætning pr. kategori')
plt.xlabel('Produktkategori')
plt.ylabel('Omsætning (DKK)')
plt.xticks(rotation=45)
plt.show()

```

```python
# 6. Top 10 produkter efter antal solgt
top_products = merged.groupby('product_name')['quantity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))

sns.barplot(
    x=top_products.values,
    y=top_products.index,
    hue=top_products.index,      # tilføjet hue
    palette="Greens_d",
    legend=False                 # fjern legend
)

plt.title('Top 10 produkter efter antal solgt')
plt.xlabel('Antal solgt')
plt.ylabel('Produktnavn')
plt.show()


```

```python

# 7. Average delivery time per city
merged['order_date'] = pd.to_datetime(merged['order_date'])
merged['shipping_date'] = pd.to_datetime(merged['shipping_date'])
merged['delivery_days'] = (merged['shipping_date'] - merged['order_date']).dt.days
avg_delivery = merged.groupby('shipping_city')['delivery_days'].mean().sort_values()
plt.figure(figsize=(8,5))

sns.barplot(
    x=avg_delivery.values,
    y=avg_delivery.index,
    hue=avg_delivery.index,
    palette="Oranges_d",
    legend=False
)
plt.title('Gennemsnitlig leveringstid pr. by')
plt.xlabel('Leveringstid (dage)')
plt.ylabel('By')
plt.show()
```

```python

# 8. Monthly sales trend
merged['month'] = merged['order_date'].dt.to_period('M')
monthly_sales = merged.groupby('month')['price'].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_sales.index.astype(str), y=monthly_sales.values, marker='o', color='blue')
plt.title('Månedlig salgsudvikling')
plt.xlabel('Måned')
plt.ylabel('Omsætning (DKK)')
plt.xticks(rotation=45)
plt.show()

```
