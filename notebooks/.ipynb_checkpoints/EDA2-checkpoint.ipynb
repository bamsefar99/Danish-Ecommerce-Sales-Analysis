```python
# Danish E-commerce: Exploratory Data Analysis (EDA2)
## 1. Importér pakker
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Indlæs data
products = pd.read_csv('../data/Products.csv')
orders = pd.read_csv('../data/Orders.csv')
```

```python
# Tilføj årskolonne

orders['order_date'] = pd.to_datetime(orders['order_date'])
orders = orders[orders['order_date'] < '2026-01-01']  # Ekskluder januar 2026 de de kun er til den 13 jan der er data til 
orders['år'] = orders['order_date'].dt.year
```

Første kig på data

```python
print(products.head())
print(orders.head())
print(products.info())
print(orders.info())
```

```python
# Merge for at få kategori på ordrer
merged = orders.merge(products[['product_id', 'product_category']], on='product_id')
oms_pr_kat = merged.groupby('product_category')['price'].sum().sort_values(ascending=False)
oms_pr_kat.plot(kind='bar', title='Omsætning pr. produktkategori', ylabel='DKK')
plt.show()
```

```python
top10 = merged.groupby('product_id')['price'].sum().sort_values(ascending=False).head(10)
top10.plot(kind='bar', title='Top 10 produkter (omsætning)', ylabel='DKK')
plt.show()

```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['month'] = orders['order_date'].dt.month
salg_pr_måned = orders.groupby('month')['price'].sum()
salg_pr_måned.plot(kind='line', marker='o', title='Sæsonvariation i omsætning', ylabel='DKK')
plt.show()
```

```python
orders['shipping_date'] = pd.to_datetime(orders['shipping_date'])
orders['delivery_days'] = (orders['shipping_date'] - orders['order_date']).dt.days
levering_pr_by = orders.groupby('shipping_city')['delivery_days'].mean().sort_values(ascending=False)
levering_pr_by.plot(kind='bar', title='Gennemsnitlig leveringstid pr. by', ylabel='Dage')
plt.show()
```

```python

# Boxplot for pris pr. kategori
sns.boxplot(x='product_category', y='price', data=merged)
plt.title('Prisfordeling pr. kategori')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(products['unit_price'], bins=20, kde=True)
plt.title('Fordeling af produktpriser')
plt.xlabel('Pris (DKK)')
plt.ylabel('Antal produkter')
plt.show()
```

```python
plt.figure(figsize=(8,4))
orders['payment_method'].value_counts().plot(kind='bar')
plt.title('Antal ordrer pr. betalingsmetode')
plt.xlabel('Betalingsmetode')
plt.ylabel('Antal ordrer')
plt.show()
```

```python

# Tilføj årskolonne
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['år'] = orders['order_date'].dt.year

# Gruppér og pivotér data
pm_per_year = orders.groupby(['år', 'payment_method']).size().unstack(fill_value=0)

# Plot
pm_per_year.plot(kind='bar', figsize=(10,6))
plt.title('Antal ordrer pr. betalingsmetode pr. år')
plt.xlabel('År')
plt.ylabel('Antal ordrer')
plt.legend(title='Betalingsmetode')
plt.tight_layout()
plt.show()
```

Analyse af betalingsmetoder pr. år
År 2024:

Bankoverførsel: 796 ordrer (33,0%)
Kreditkort: 823 ordrer (34,1%)
MobilePay: 795 ordrer (32,9%)
→ Mest populær: Kreditkort

År 2025:

Bankoverførsel: 817 ordrer (32,5%)
Kreditkort: 853 ordrer (33,9%)
MobilePay: 844 ordrer (33,6%)
→ Mest populær: Kreditkort

År 2026:

Bankoverførsel: 21 ordrer (29,2%)
Kreditkort: 36 ordrer (50,0%)
MobilePay: 15 ordrer (20,8%)
→ Mest populær: Kreditkort

```python
Overordnet trend

Kreditkort er den mest populære betalingsmetode i alle årene, men forskellen til MobilePay og Bankoverførsel er lille i 2024 og 2025.
MobilePay og Bankoverførsel deler resten af markedet næsten ligeligt, men Kreditkort vokser svagt over tid.
I 2026 er Kreditkort markant mere populær, men datagrundlaget for 2026 er meget mindre (kun 72 ordrer).
```

```python
orders['month'] = pd.to_datetime(orders['order_date']).dt.month
pivot = orders.pivot_table(index='shipping_city', columns='month', values='order_id', aggfunc='count')
plt.figure(figsize=(12,6))
sns.heatmap(pivot, cmap='Blues', annot=True, fmt='.0f')
plt.title('Antal ordrer pr. by og måned')
plt.xlabel('Måned')
plt.ylabel('By')
plt.show()
```
